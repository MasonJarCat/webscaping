```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time
import os
from dotenv import load_dotenv

# Load environment variables from a .env file
load_dotenv()
password = os.getenv("PASS")

# Set up Chrome options to make the browser run in headless mode (optional)
chrome_options = Options()
chrome_options.add_argument(
    "--headless"
)  # Comment out if you want to see the browser actions


driver = webdriver.Chrome()

driver.implicitly_wait(5)
# Open the target page
url = "https://networking.invisso.org/event/abs-east-2024/people/RXZlbnRWaWV3Xzg5MTA1OA=="  # Replace with the actual URL
driver.get(url)
time.sleep(1)
try:
    refuse_all_button = driver.find_element(
        By.CSS_SELECTOR, '[data-hook="cookie-banner-refuse-all-button"]'
    )
    refuse_all_button.click()
    print("Clicked the 'Refuse all' button.")
except Exception as e:
    print(f"Failed to click 'Refuse all' button: {e}")

time.sleep(1)
try:
    login_button = driver.find_element(By.XPATH, '//button[@aria-label="Log in"]')
    login_button.click()
    print("Clicked the 'Log in' button.")
except Exception as e:
    print(f"Failed to click 'Log in' button: {e}")
time.sleep(1)
try:
    # Locate the input field by its ID
    email_input = driver.find_element(By.ID, "lookup-email-input-id")

    # Input the email address into the field
    email_input.send_keys(
        "ergkys@revl.net"
    )  # Replace with the actual email you want to use

    print("Email input successful.")

except Exception as e:
    print(f"Failed to input email: {e}")
time.sleep(1)
try:
    # Locate the "Continue" button using the data-hook attribute
    continue_button = driver.find_element(
        By.CSS_SELECTOR, '[data-hook="login-email-submit"]'
    )

    # Click the "Continue" button
    continue_button.click()

    print("Continue button clicked successfully.")

except Exception as e:
    print(f"Failed to click the Continue button: {e}")

time.sleep(1)
try:
    # Locate the password input field by its ID
    password_input = driver.find_element(
        By.ID, "login-with-email-and-password-password-id"
    )

    # Input the password into the field
    password_input.send_keys(
        password
    )  # Replace with the actual password you want to use

    print("Password input successful.")

except Exception as e:
    print(f"Failed to input password: {e}")

time.sleep(1)
try:
    # Locate the login button using the data-hook attribute
    login_button = driver.find_element(
        By.CSS_SELECTOR, '[data-hook="login-password-submit"]'
    )

    # Click the login button
    login_button.click()

    print("Login button clicked successfully.")

except Exception as e:
    print(f"Failed to click the login button: {e}")

time.sleep(10)


def extract_person_info():
    try:
        time.sleep(0.1)
        try:
            # Extract company name
            company_name = driver.find_element(
                By.CSS_SELECTOR, "div.sc-45c084b0-2.jsVWbR"
            ).text
        except Exception as e:
            print(f"Failed to extract company name: {e}")
            company_name = None

        try:
            # Extract classification
            classification = driver.find_element(
                By.XPATH, '//div[text()="Classification"]/following-sibling::div'
            ).text
        except Exception as e:
            print(f"Failed to extract classification: {e}")
            classification = None

        # Extract interest areas
        try:
            interest_areas_elements = driver.find_elements(
                By.XPATH, '//div[text()="Interest Areas"]/following-sibling::div//span'
            )
            interest_areas = ", ".join([elem.text for elem in interest_areas_elements])
        except Exception as e:
            print(f"Failed to extract interest areas: {e}")
            interest_areas = None
            # Extract person's name
        try:
            person_name = driver.find_element(
                By.CSS_SELECTOR, "h2.sc-f3b80a80-1.hbvnEe"
            ).text
        except Exception as e:
            print(f"Failed to extract person name: {e}")
            person_name = None
        # Extract role of the person
        try:
            role = driver.find_element(By.CSS_SELECTOR, "h4.sc-f3b80a80-2.gqcPin").text
        except Exception as e:
            print(f"Failed to extract role: {e}")
            role = None

        # Return the extracted data
        return {
            "Person Name": person_name,
            "Role": role,
            "Company Name": company_name,
            "Classification": classification,
            "Interest Areas": interest_areas,
            "url": url,
        }

    except Exception as e:
        print(f"Error extracting data: {e}")
        return None


people = []
# Extract data from the page and print the result
with open("links2.txt", "r") as file:
    urls = file.readlines()
for url in urls:
    driver.get(url)
    time.sleep(1)
    person_info = extract_person_info()
    if person_info:
        people.append(person_info)
        print(person_info)
    else:
        print("Failed to extract person info.")
        break


df1 = pd.DataFrame(people)
df1.to_csv("people2.csv", index=False)

driver.quit()
# Close the browser
```
